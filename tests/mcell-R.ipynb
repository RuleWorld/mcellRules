```python
import pandas
totaldataset = pandas.read_hdf('{0}DB.h5'.format('blbr'))

```

```python
mdldataset = totaldataset.reset_index()
rmcell = mdldataset.query('Origin == "R-MCell"')

nfsim = mdldataset.query('Origin == "NFsim"')
rts = set(rmcell['Seconds'])
nts = set(nfsim['Seconds']).intersection(rts)

mdldataset = mdldataset[mdldataset.Seconds.isin(nts)]
rowdicts = []
for l, d in mdldataset.groupby("Seconds iteration Origin Observable".split()):
    d = {"Seconds": l[0], "iteration": l[1], "Origin": l[2], "Observable":l[3], "Value":list(d.Value)[0]}
    rowdicts.append(d)                               

corrected = pandas.DataFrame.from_dict(rowdicts)

corrected = mdldataset


corrected = corrected.query('Observable != "Rtot" & Observable != "Ltot"')
```

```python
%%output holomap='gif' fps=5

def fix_lims(plot, el):
    ax = plot.handles['axis']
    ax.set_xlim(0, 450)
    ax.set_ylim(0, 0.1)
    ax.set_aspect(50/0.01)
    ax.legend()
    
```

```python
%%opts Overlay [final_hooks=[fix_lims]]
#gridspace = hv.GridSpace(kdims=['Observable'])
hv.Distribution.fig_rcparams={'axes.labelsize':40, 'axes.titlesize':40}
gridspace = []

#mkde = hv.HoloMap(kdims=['Seconds'])
#nkde = hv.HoloMap(kdims=['Seconds'])
timerows = list(set(mdldataset['Seconds']))
timerows = sorted(timerows)
timerows = timerows[::5]
for obs in set(corrected.Observable):
    minx = min(corrected.query("Observable == '{0}'".format(obs)).Value)
    maxx = max(corrected.query("Observable == '{0}'".format(obs)).Value)
    print obs, minx,maxx
    kde1 = hv.HoloMap(kdims=['Seconds']).relabel(label=obs,depth=3) 
    kde2 = hv.HoloMap(kdims=['Seconds']).relabel(label=obs,depth=3) 
    #kde = hv.HoloMap(kdims=['Seconds']).relabel(label=obs,depth=3) 
    for te in timerows:
        value = corrected.query("Seconds == {0} & Observable == '{1}'".format(te, obs))
        ts1 = hv.Distribution(value.query("Origin == 'NFsim'").Value, extents=(minx,0,maxx,0.15), label="NFsim").relabel(label=obs,depth=3)
        ts2 = hv.Distribution(value.query("Origin == 'R-MCell'").Value, extents=(minx,0,maxx,0.15), label="MCell-R").relabel(label=obs,depth=3)
        #ndoverlay = hv.NdOverlay(ts)

        kde1[te]  = ts1
        kde2[te]  = ts2
        #kde[te] = hv.Distribution(value.query("Origin == '{0}'".format("R-MCell")).Value, , label="R-MCell")
    kde = hv.HoloMap({'MCell-R':kde2,'NFsim':kde1},kdims=['Seconds']).collate()
    
    #gridspace[obs] = kde
    gridspace.append(kde1*kde2)

    

```

```python
%%output holomap='gif' fps=5

def fix_lims(plot, el):
    ax = plot.handles['axis']
    ax.set_xlim(0, 900)
    ax.set_ylim(0, 0.02)
    ax.set_aspect(900/0.02)
    ax.legend()
    
```

```python
%%opts Overlay [final_hooks=[fix_lims]]
def myRandom(mult, offset):
    return (mult) * np.random.randn(500) + offset

hm1 = hv.HoloMap(kdims=['Seconds'])
hm2 = hv.HoloMap(kdims=['Seconds'])
hm3 = hv.HoloMap(kdims=['Seconds'])

final1 = hv.HoloMap(kdims=['Seconds'])
final2 = hv.HoloMap(kdims=['Seconds'])
final3 = hv.HoloMap(kdims=['Seconds'])
for t in range(1,20):
    d1 = hv.Distribution(myRandom(25,t*10+450), label='Blue', extents=(0,0,900,0.02))
    d2 = hv.Distribution(myRandom(45,t*10+540), label='Red', extents=(0,0,900,0.02))                           
    d3 = hv.Distribution(myRandom(55,t*10+590), label='Green', extents=(0,0,900,0.02))
    ndov = hv.NdOverlay({'blue':d1, 'red':d2, 'green':d3})
    hm1[t] = d1
    hm2[t] = d2
    hm3[t] = d3
    final1[t] = d1*d2*d3
    final3[t] = ndov
final2 = hm1*hm2 *hm3

hm1 + final1 + final2 + final3 + ndov
                             
```
